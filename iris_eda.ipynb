```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#Thiết lập phong cách giao diện chung cho biểu đồ
sns.set_theme(style="whitegrid")

iris = sns.load_dataset('iris')

display(iris.head())
```

```python
#Kiểm tra tính cân bằng của dữ liệu
species_count = iris['species'].value_counts()
print(species_count)
```

```python
#Vẽ biểu đồ histograms xem phân phối từng đặc trưng
iris.hist(figsize=(10, 8), bins=20, edgecolor='black')
plt.suptitle("Histograms of Iris Features", fontsize=16)
plt.show()

#Tạo Box plots tìm kiếm giá trị ngoại lệ (outliers)
plt.figure(figsize=(10, 6))
sns.boxplot(data=iris, orient='h', palette="Set2")
plt.title("Box plots of Iris Features", fontsize=16)
plt.show()
```

```python
#Pairplot so sánh all cặp đặc trưng
sns.pairplot(iris, hue="species", markers=["o", "s", "D"], palette="husl")
plt.suptitle("Pairplot of Iris Dataset", y=1.02, fontsize=16)
plt.show()

#=> setposa tách biệt - linear separable - có thể phân tách bằng đường thẳng
```

```python
#Ma trận tương quan (Correlation Matrix & HeatMap)
#Chỉ tính tương quan trên các cột số (bỏ cột species dạng chữ)
corr_matrix = iris.drop(columns='species').corr()

#hiển thị bằng heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix Heatmap", fontsize=16)
plt.show()
```

### Summary / Kết luận
Qua quá trình phân tích (EDA), chúng ta có thể rút ra một số kết luận sau:
1. **Tính tách biệt:** Loài **Setosa** hoàn toàn tách biệt khỏi hai loài còn lại (Versicolor và Virginica) dựa trên kích thước cánh hoa (Petal length và Petal width).
2. **Độ tương quan:** Có sự tương quan dương rất mạnh (gần 0.96) giữa chiều dài cánh hoa (Petal length) và chiều rộng cánh hoa (Petal width).
3. **Mức độ khó phân loại:** Loài Versicolor và Virginica có một chút chồng chéo lên nhau về mặt kích thước, tuy nhiên Virginica nhìn chung có kích thước lớn hơn ở hầu hết các đặc trưng.
